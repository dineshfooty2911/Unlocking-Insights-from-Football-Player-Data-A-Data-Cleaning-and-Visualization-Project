# Data cleaning
When working with datasets, null or missing values can be a common issue that needs to be addressed.
Null values in the dataset were managed by replacing some null values with zeros and dropping rows with other null values to ensure the integrity of the data was maintained. This step  ensured that missing data did not impact the analysis and was a important step in preparing the dataset for further analysis.

```python
ndf.duplicated().sum()
```

Using the sum fuction we can determind how many Null vales are present in each Column

```python
ndf.isnull().sum()
```

#### By visualizing the missing data using a heatmap, we can quickly identify which parts of the dataset are missing data and how much data is missing.

```python
sns.heatmap(ndf.isnull(),cmap='Reds')
```

The percentage of missing values are determined|

```python

ndf.isnull().sum()/(len(ndf))*100
```

####  When the Missing or null values is very less we can fill the Null vales by zero 

```python
ndf['value_eur'].fillna(0, inplace=True)
ndf['wage_eur'].fillna(0, inplace=True)
ndf['club_joined'].fillna("Unknown", inplace=True)
ndf['club_name'].fillna("Unknown", inplace=True)
ndf['club_position'].fillna("Unknown", inplace=True)
```

####  When there are large clusters of Null value we should not replace by zero which will affect the Data integrity and accuracy of analysis , so we are dropping rows with other null values

```python
ndf.dropna(inplace=True)
```

```python
ndf.isnull().sum()
```

#### Now when we check the heat map for Null values it looks Perefect and the Null values are handled

```python
sns.heatmap(ndf.isnull())
```

# The Rador plot for the first player is plotted to analysze his overall abilites

```python
# Messi rador plot needs to be customized
from math import pi
rdf=nndf[['shooting','passing','dribbling','defending','physic']]
categories=list(rdf)[0:]
N = len(categories)

values=rdf.loc[0].values.flatten().tolist()
values += values[:1]

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1],color='Black',size=8)
ax.set_xticklabels(categories, size=10)

ax.set_rlabel_position(0)
plt.yticks([10,20,30,40,50,60,70,80,90], color="Blue", size=7)
plt.ylim(0,100)
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Messi")
ax.fill(angles, values, 'r', alpha=0.2)
plt.show()
```

### The rador Plot to compare two players based on their abilities

```python

rdf=nndf[['shooting','passing','dribbling','defending','physic']]
categories=list(rdf)[0:]
N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1], categories)

ax.set_rlabel_position(0)
plt.yticks([10,20,30,40,50,60,70,80,90], ["10","20","30","40","50","60","70","80","90"], color="Blue", size=7)
plt.ylim(0,100)

values=rdf.loc[0].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Messi")
ax.fill(angles, values, 'b', alpha=0.1)

values=rdf.loc[2].values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Roandlo")
ax.fill(angles, values, 'r', alpha=0.1)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```

### The chart displays the count of players from each club

```python
hdf=nndf[['club_name']].head(100)
fig, ax = plt.subplots(figsize=(10, 5))
sns.set_palette("pastel")
sns.histplot(data=hdf, x="club_name",ax=ax,)
plt.xticks(rotation=90)
ax.tick_params(axis='x', colors='red')

ax = plt.gca()

# For each bar, get the height and the x-position
for i, bar in enumerate(ax.patches):
    height = bar.get_height()
    x_position = bar.get_x() + bar.get_width() / 2

    # Annotate the bar with its height
    ax.annotate(f"{int(height)}", xy=(x_position, height), xytext=(0, 3), 
                textcoords="offset points", ha='center', va='bottom')

plt.show()
```
